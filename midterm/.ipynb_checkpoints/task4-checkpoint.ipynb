```python
from scipy import ndimage 
from scipy import ndimage as ndi 
import numpy as np 
import matplotlib.image as img 
import matplotlib.pyplot as plt 
from skimage import color 
from skimage import io 
from skimage import util 
from skimage import feature 
import matplotlib.image as pltimg 

fig = plt.figure(figsize=(50,40))
img = io.imread('32409.jpg', as_grey=True) 
canny_pic = feature.canny(img) 
sx = ndi.sobel(img, axis = 0, mode = 'constant') 
sy = ndi.sobel(img, axis = 1, mode = 'constant') 
sobel_pic = np.hypot(sy,sx) 

plt.subplot(221) 
plt.title('Original') 
plt.imshow(img,cmap='gray') 

plt.subplot(223) 
plt.title('Canny') 
plt.imshow(canny_pic,cmap='gray') 

plt.subplot(224) 
plt.title('Sobel') 
plt.imshow(sobel_pic,cmap='gray') 
plt.show()
```
