```python
from scipy import ndimage 
import numpy as np 
import matplotlib.image as img 
import matplotlib.pyplot as plt 
from skimage import color 
from skimage import io 
from skimage import util 

fig = plt.figure(figsize=(35,35)) 

import matplotlib.image as pltimg 
image = io.imread('32409.png', as_grey=True) 
image = util.random_noise(image, mode='s&p', seed=None, clip=True) 

plt.subplot(321) 
plt.title('Original') 
plt.imshow(image,cmap='gray') 

A = np.array([[1.*1/6,2.*1/6,1.*1/6],
              [2.*1/6,4.*1/6,2.*1/6],
              [1.*1/6,2.*1/6,1.*1/6]]) 
im_1 = ndimage.convolve(image, A, mode='nearest', cval=0.0) 

plt.subplot(322) 
plt.title('Filtr A') 
plt.imshow(im_1,cmap='gray') 

B = np.array([[0.*1/5,1.*1/5,0.*1/5],
              [1.*1/5,1.*1/5,1.*1/5],
              [0.*1/5,1.*1/5,0.*1/5]]) 
im_2 = ndimage.convolve(image, B, mode='nearest', cval=0.0) 

plt.subplot(323) 
plt.title('Filtr B') 
plt.imshow(im_2,cmap='gray') 

#Median filter
im_3 = ndimage.median_filter(image, 3) 

plt.subplot(324) 
plt.title('Median filter') 
plt.imshow(im_3,cmap='gray') 

C = np.array([[-1.,-2.,-1.],
              [-2.,12.,-2.],
              [-1.,-2.,-1.]]) 
im_4 = ndimage.convolve(image, C, mode='nearest', cval=0.0) 
 
plt.subplot(325) 
plt.title('Filtr D') 
plt.imshow(im_4,cmap='gray') 

D = np.array([[-1.,-2.,-1.],
              [0.,0.,0.],
              [1.,2.,1.]]) 
im_5 = ndimage.convolve(image, D, mode='nearest', cval=0.0) 

plt.subplot(326) 
plt.title('Filtr E') 
plt.imshow(im_5,cmap='gray') 
plt.show()
```
