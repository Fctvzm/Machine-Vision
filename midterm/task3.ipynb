```python
from scipy import ndimage 
import numpy as np 
import matplotlib.image as img 
import matplotlib.pyplot as plt 
from skimage import color 
from skimage import io 
from skimage import util 
import matplotlib.image as pltimg 

image = io.imread('32409.png', as_grey=True) 
image_s_p = util.random_noise(image, mode='s&p', seed=None, clip=True) 

fig = plt.figure(figsize=(50,40)) 
plt.subplot(221) 
plt.title('Original') 
plt.imshow(image,cmap='gray') 

#Salt and pepper
plt.subplot(222) 
plt.title('Salt and pepper') 
plt.imshow(image_s_p,cmap='gray') 

plt.subplot(223) 
plt.title('Original') 
plt.imshow(image,cmap='gray') 

#median filter
flt = ndimage.median_filter(image, 3) 

plt.subplot(224) 
plt.title('Median Filter') 
plt.imshow(flt,cmap='gray') 

plt.show()
```
